```python
import re
import csv
import pandas as pd
class invalid(Exception):  
    pass
print("1)if you are a new user then register write register")
print("2)if you are a existed user the write login")
user=input("select register or login or forgot password:") # User selection
a=open("/content/data2.csv","a")# csv file to store the data
a.write('"email","password"\n')
if user=="register": # If user chooses register
    while True:
        email=input("put your email id or username:")
        try:
            if "@" not in email or "." not in email: # "@" and "." should be in email id
                raise invalid
            elif email[0] in "~`!@#$%^&*-_+=/\'." or email[0] in "0123456789": # Condition for special and number
                raise invalid
            elif email[0]=="@" or email[0]==".": # Start of email shoul not be with "@" and "."
                raise invalid
            elif len(email[email.index("@"):email.index(".")])<4: # "@" and "." should not be adjacent
                raise invalid
            elif email.count("@")!=1 and email.count(".")!=1: # "@" and "." should not be more than 1
                raise invalid
            break
        except invalid:
            print("email does not meet email format standards, try again")

    while True:
        try:
            password=input("create your password(Must have minimum one special character,one digit,one uppercase,one lowercase character):")
            if len(password)<5 and len(password)>16: # Password length should be between 5 to 16
                raise invalid
            b=re.findall("\W",password) # There should at least be one special character
            if len(b)==0:
                raise invalid
            c=re.findall("\d",password) # There should at least be one number
            if len(c)==0:
                raise invalid
            d=[]
            e=[]
            for i in password:
                if i.isupper()==True: # There should at least one upper character
                    d.append("true")
                else:
                    e.append("false")
            if len(d)==0:
                raise invalid
            f=[]
            g=[]
            for i in password:
                if i.islower()==True: # There should at least be one lower character
                    f.append("true")
                else:
                    g.append("false")
            if len(f)==0:
                raise invalid
            break
        except:
            print("this password does not meet password format standards, try again")
    a.write('"{}","{}"\n'.format(email,password)) # All condtions are satisfied then store email and password in the file
elif user=="login": # If user chooses to login(i.e user is existing user)
    email=input("put your existing email id or username:")
    password=input("password:")
    a=open("/content/data2.csv","r")
    b=csv.reader(a,delimiter=",")
    h=[]
    l=[]
    for i in b:
        for j in range(0,len(i)):
            if email==i[j] or password==i[j]:
                h.append(1)
            else:
                l.append(0)
    if len(h)==0:
        print("not existing details, go for registration")
    else:
        print("existing account")
elif user=="forgot password": # If user chooses forgot password option
    choose=input("select retrieve password or create new password:") # Either user wants to retrieve old password or create new password
    if choose=="retrieve password": # If user chooses to retrieve password
        email=input("put your existing username or email id:") # Existing email id
        a=open("/content/data2.csv","r")
        b=csv.reader(a,delimiter=",")
        m=[]
        for i in b:
            for j in range(0,len(i)):
                if email==i[j]:
                    m.append(1)
                    a=i[j+1]
        if len(m)==0:
            print("email id does not exist, go for registration")
        else:
            print(a)
    elif choose=="create new password": # If user chooses to create a new password
        email=input("put your existing  username or email id:")
        a=open("/content/data2.csv","r")
        b=csv.reader(a,delimiter=",")
        n=[]
        for i in b:
            for j in range(0,len(i)):
                if email==i[j]:
                    n.append(1)
        if len(n)==0:
            print("email id does not exist, go for registation")
        else:
            while True:
                try:
                    password=input("create a new password(Must have minimum one special character,one digit,one uppercase,one lowercase character):")
                    if len(password)<5 and len(password)>16:
                        raise invalid
                    b=re.findall("\W",password)
                    if len(b)==0:
                        raise invalid
                    c=re.findall("\d",password)
                    if len(c)==0:
                        raise invalid
                    d=[]
                    e=[]
                    for i in password:
                        if i.isupper()==True:
                            d.append("true")
                        else:
                            e.append("false")
                    if len(d)==0:
                        raise invalid
                    f=[]
                    g=[]
                    for i in password:
                        if i.islower()==True:
                            f.append("true")
                        else:
                            g.append("false")
                    if len(f)==0:
                        raise invalid
                    break
                except:
                    print("invalid password, try again")
            df=pd.read_csv("/content/data2.csv")
            o=df[df["email"]==email].index.values
            df["password"][o]=password
            df.to_csv("/content/data2.csv")
            
```
